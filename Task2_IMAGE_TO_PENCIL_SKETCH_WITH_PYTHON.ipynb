```python
from google.colab.patches import cv2_imshow
import cv2
```

```python
image = cv2.imread("jake.jpg")
cv2_imshow(image)
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(image_gray)
```

```python
image_invert = cv2.bitwise_not(image_gray)
cv2_imshow(image_invert)
```

```python
image_smoothing = cv2.GaussianBlur(image_invert, (21, 21),sigmaX=0, sigmaY=0)
cv2_imshow(image_smoothing)
```

```python
def final_sketch(x, y):
    return cv2.divide(x, 255 - y, scale=256)
final_image = final_sketch(image_gray, image_smoothing)
cv2_imshow(final_image)
```
